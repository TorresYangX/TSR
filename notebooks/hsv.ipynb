```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def log2_sat(x):
    return np.log2(x + 1)

def sqrt_sat(x):
    return np.sqrt(x)
```

```python
img = cv2.imread('tmp/0.jpg')

plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR))

plt.imsave('tmp/original.png', cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
```

```python
mv_g = img.copy()
mv_g[..., 1] = 0
# mv_g[..., 2]
plt.imshow(cv2.cvtColor(mv_g, cv2.COLOR_RGB2BGR))
```

```python
hsv = cv2.cvtColor(mv_g, cv2.COLOR_BGR2HSV)

# print(hsv)
hsv[..., 2]  =  sqrt_sat(hsv[..., 2] / 255) * 255

# print(hsv)
result = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
plt.imshow(result)
```

```python
plt.imsave('tmp/result.png', result)
```
